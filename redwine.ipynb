```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
print("working dir: ", os.getcwd())
# load the .xlsx file here
df = pd.read_excel(r"C:\Users\mridr\anaconda_projects\cf39e985-5dd2-4b16-b9ff-172704c071d3\winedata\winequality-red.xlsx", engine='openpyxl', header=1)

# Show the first data frame
display(df)

# Create a new DataFrame without duplicates
df_no_duplicates = df.drop_duplicates()

# Show the new DataFrame
display(df_no_duplicates)

for column in df.columns:
    plt.figure(figsize=(4, 6))
    plt.boxplot(df[column].dropna())
    plt.title(f"Boxplot of {column}")
    plt.ylabel(column)
    plt.grid(True) #git p
    plt.tight_layout()
    plt.show()


```

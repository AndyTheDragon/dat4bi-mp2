```python
import pandas as pd

white_wine = pd.read_excel('data/winequality-white.xlsx')
```

```python
white = white_wine.copy()
white.columns = white.iloc[0]
white = white[1:]
white['type'] = 'white'
```

```python
# Convert all columns except 'type' to numeric, errors='coerce' will turn non-convertible values to NaN
for col in white.columns:
    if col != 'type':
        white[col] = pd.to_numeric(white[col], errors='coerce')
```

```python
white.reset_index(drop=True, inplace=True)
```

```python
duplicates = white[white.duplicated()]
duplicates
```

```python
white_no_duplicates = white.drop_duplicates()
```

```python
white_no_duplicates.reset_index(drop=True, inplace=True)
white_no_duplicates
```

```python
white
```

```python
ranges = {}
for col in white.columns:
    min_val = white[col].min()
    max_val = white[col].max()
    ranges[col] = (min_val, max_val)

ranges

```

```python
import matplotlib.pyplot as plt

# Select only numeric columns for boxplot
numeric_cols = white.select_dtypes(include='number').columns

plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
white[numeric_cols].boxplot(rot=90)
plt.title('White Wine (with duplicates)')
plt.ylabel('Value')

plt.subplot(1, 2, 2)
white_no_duplicates[numeric_cols].boxplot(rot=90)
plt.title('White Wine (no duplicates)')
plt.ylabel('Value')

plt.tight_layout()
plt.show()
```

Vi kan udlede fra disse diagrammer at boxplot ikke giver mening at bruge ift dupliketter vs ikke dupliketter.

```python
white_summary = white.describe(include='all').transpose().drop(columns=['unique', 'top', 'freq'], errors='ignore')
white_no_duplicates_summary = white_no_duplicates.describe(include='all').transpose().drop(columns=['unique', 'top', 'freq'], errors='ignore')

print("White Wine Summary (with duplicates):")
display(white_summary)
print("\nWhite Wine Summary (no duplicates):")
display(white_no_duplicates_summary)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation
corr = white_no_duplicates[numeric_cols].corr()

# Heatmap (unchanged)
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of White Wine Features (No Duplicates)')
plt.show()

# Upper-triangle mask to drop self-correlation and duplicates
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)

# Take absolute correlations, stack, name axes and series to avoid name collisions
high_corr = (
    corr.where(~mask)             # keep lower triangle only
        .abs()
        .stack()                  # -> Series with MultiIndex (i, j)
        .rename('Correlation')    # name the values column
        .rename_axis(['Feature 1', 'Feature 2'])  # name the index levels
        .reset_index()            # now safe: no conflicting names
        .sort_values('Correlation', ascending=False)
)

display(high_corr)
```
